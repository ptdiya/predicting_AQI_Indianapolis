```python
import numpy as np
import holidays
from workalendar.usa import UnitedStates

final_features['date'] = pd.to_datetime(final_features['date'])

us_holidays = holidays.US()
from datetime import timedelta

def is_observed_holiday(date):
    if date in us_holidays:
        return 1
    
    if (date.weekday() == 0) and (date - timedelta(days=1)) in us_holidays:  
        return 1
    
    if (date.weekday() == 4) and (date + timedelta(days=1)) in us_holidays:
        return 1
    
    return 0

final_features['is_holiday'] = final_features['date'].apply(lambda x: is_observed_holiday(x))

cal = UnitedStates()
final_features['is_school_break'] = final_features['date'].apply(lambda x: 1 if cal.is_working_day(x) == False else 0)

from datetime import date

def is_between(date, start_month, start_day, end_month, end_day):
    start = date.replace(month=start_month, day=start_day)
    end = date.replace(month=end_month, day=end_day)
    if start <= end:
        return start <= date <= end
    else: 
        return date >= start or date <= end

final_features['is_winter_break'] = final_features['date'].apply(lambda x: 1 if is_between(x, 12, 20, 1, 5) else 0)
final_features['is_summer_break'] = final_features['date'].apply(lambda x: 1 if is_between(x, 6, 1, 8, 31) else 0)
final_features['is_weekend'] = final_features['date'].dt.dayofweek.isin([5, 6]).astype(int)
final_features['is_break'] = final_features['date'].apply(lambda x: 1 if not cal.is_working_day(x) else 0)

def is_long_weekend(date):
    friday = date - timedelta(days=(date.weekday() - 4) % 7)
    monday = date + timedelta(days=(7 - date.weekday()) % 7)
    
    if not cal.is_working_day(friday) or not cal.is_working_day(monday):
        return 1 if date in [friday, friday + timedelta(days=1), friday + timedelta(days=2),
                             monday - timedelta(days=2), monday - timedelta(days=1), monday] else 0
    else:
        return 0 
        
final_features['is_long_weekend'] = final_features['date'].apply(is_long_weekend)


def days_to_break(date):
    current_year = date.year
    next_year = current_year + 1
    summer_start = date.replace(month=6, day=1)
    summer_end = date.replace(month=8, day=31)
    winter_start = date.replace(month=12, day=20)
    winter_end = date.replace(year=next_year, month=1, day=5)

    if date < summer_start:
        days_to_summer = (summer_start - date).days
    elif date > summer_end:
        days_to_summer = (summer_start.replace(year=next_year) - date).days
    else:
        days_to_summer = 0  

    if date < winter_start:
        days_to_winter = (winter_start - date).days
    elif date > winter_end:
        days_to_winter = (winter_start.replace(year=next_year) - date).days
    else:
        days_to_winter = 0  

    return min(days_to_summer, days_to_winter)

final_features['days_to_break'] = final_features['date'].apply(days_to_break)
final_features['break_proximity'] = np.exp(-final_features['days_to_break'] / 30)  # exponential decay
```

```python
def add_cyclical_features(df, col_name, period):
    df[f'{col_name}_sin'] = np.sin(2 * np.pi * df[col_name] / period)
    df[f'{col_name}_cos'] = np.cos(2 * np.pi * df[col_name] / period)
    return df

final_features['date'] = pd.to_datetime(final_features['date'])

#cyclical components
final_features['month'] = final_features['date'].dt.month
final_features['quarter'] = final_features['date'].dt.quarter
final_features['four_month'] = (final_features['month'] - 1) // 4 + 1
final_features['half_year'] = (final_features['month'] - 1) // 6 + 1

final_features = add_cyclical_features(final_features, 'month', 12)
final_features = add_cyclical_features(final_features, 'week_number', 52)
final_features = add_cyclical_features(final_features, 'quarter', 4)
final_features = add_cyclical_features(final_features, 'four_month', 3)
final_features = add_cyclical_features(final_features, 'half_year', 2)

final_features.head()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

X_train_numeric = X_train.select_dtypes(include=[float, int])
X_val_numeric = X_val.select_dtypes(include=[float, int])

X_train_tensor = torch.tensor(X_train_numeric.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)  
X_val_tensor = torch.tensor(X_val_numeric.values, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val.values, dtype=torch.float32)  
```

```python
from torch.utils.data import DataLoader, TensorDataset
```

```python
y_train_labels = y_train_tensor
y_val_labels = y_val_tensor

sequence_length = 5

def create_sequences(X, y_labels, seq_length):
    sequences = []
    targets = []
    for i in range(seq_length, len(X)):
        X_seq = X[i-seq_length:i]  
        y_seq = y_labels[i]        
        sequences.append(X_seq)
        targets.append(y_seq)
    return torch.stack(sequences), torch.stack(targets).long()
```

```python
X_train_sequences, y_train_sequences = create_sequences(X_train_tensor, y_train_labels, sequence_length)
X_val_sequences, y_val_sequences = create_sequences(X_val_tensor, y_val_labels, sequence_length)
```

```python
class TimeSeriesDataset(torch.utils.data.Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
batch_size = 32
train_dataset = TimeSeriesDataset(X_train_sequences, y_train_sequences)
val_dataset = TimeSeriesDataset(X_val_sequences, y_val_sequences)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
```

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes, dropout=0.5):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(
            input_size, hidden_size, num_layers=num_layers, 
            batch_first=True, dropout=dropout
        )
        self.fc1 = nn.Linear(hidden_size, hidden_size // 2)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(dropout)
        self.fc2 = nn.Linear(hidden_size // 2, num_classes)
    def forward(self, x):
        out, _ = self.lstm(x)
        out = out[:, -1, :]
        out = self.fc1(out)
        out = self.relu(out)
        out = self.dropout(out)
        out = self.fc2(out)
        return out
```

```python
input_size = X_train_sequences.shape[2]
hidden_size = 512
num_layers = 10 
num_epochs = 70
num_classes = 5
```

```python
model = LSTMModel(input_size, hidden_size, num_layers, num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=0.0001, weight_decay=1e-7)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

```python
# scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10)

# scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=5)

scheduler = optim.lr_scheduler.OneCycleLR(
    optimizer, max_lr=0.001, steps_per_epoch=len(train_loader), epochs=num_epochs
)

# scheduler = optim.lr_scheduler.CyclicLR(
#     optimizer, base_lr=0.0001, max_lr=0.01, step_size_up=5, mode='triangular2'
# )

# optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-5)
```

```python
early_stop_patience = 10
best_val_loss = float('inf')
epochs_no_improve = 0

for epoch in range(num_epochs):
    model.train()
    train_loss = 0
    correct = 0
    total = 0

    for X_batch, y_batch in train_loader:
        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)

        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        train_loss += loss.item() * X_batch.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += y_batch.size(0)
        correct += (predicted == y_batch).sum().item()

    scheduler.step()

    train_loss /= len(train_loader.dataset)
    train_accuracy = 100 * correct / total
    
    model.eval()
    val_loss = 0
    correct = 0
    total = 0

    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            X_batch = X_batch.to(device)
            y_batch = y_batch.to(device)
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            val_loss += loss.item() * X_batch.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += y_batch.size(0)
            correct += (predicted == y_batch).sum().item()

    val_loss /= len(val_loader.dataset)
    val_accuracy = 100 * correct / total

    print(f'Epoch [{epoch+1}/{num_epochs}], '
          f'Train Loss: {train_loss:.4f}, Train Acc: {train_accuracy:.2f}%, '
          f'Val Loss: {val_loss:.4f}, Val Acc: {val_accuracy:.2f}%')

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        epochs_no_improve = 0
        best_model = model.state_dict()
    else:
        epochs_no_improve += 1
        if epochs_no_improve == early_stop_patience:
            print("Early stopping!")
            model.load_state_dict(best_model)
            break
```
